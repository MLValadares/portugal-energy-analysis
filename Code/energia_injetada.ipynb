```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Energia Injetada na Rede de Distribuição
drive.mount('/content/drive')
df = pd.read_parquet('/content/drive/MyDrive/Colab Notebooks/tese/energia-injetada-na-rede-de-distribuicao.parquet')
print(df.head())
```

```python
df['datahora'] = pd.to_datetime(df['datahora'])
df
```

```python
# hibricas
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['datahora']
  ys = series['hidrica']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('datahora', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('ano')):
  _plot_series(series, series_name, i)
  fig.legend(title='ano', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('datahora')
_ = plt.ylabel('hidrica')
```
